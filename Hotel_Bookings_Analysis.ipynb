### Data Analysis and Visualisation

```python
cancelled_perc = df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)
plt.figure(figsize=(7,7))
plt.title("Reservation status count")
plt.bar(['Not canceled','Canceled'],df['is_canceled'].value_counts(),edgecolor='k',width=0.4)
```

 Hence,62.865% of reservations are not cancelled and 37.13% are cancelled which is high

```python
plt.figure(figsize=(7,7))
ax1 = sns.countplot(x = "hotel",hue = "is_canceled",data = df,palette='pastel')
legend_labels,_ = ax1.get_legend_handles_labels()
# ax1.legend(bbox_to_anchor(1,1))
plt.title("Reservation status count")
plt.xlabel('hotel')
plt.ylabel('number of reservations')
```

Cancellation ratio is more in City Hotels than Resort Hotels

```python
resort_hotels = df[df['hotel']=='Resort Hotel']
city_hotels = df[df['hotel']=='City Hotel']
print("For Resort Hotels :")
print(resort_hotels['is_canceled'].value_counts(normalize=True))
print("For City Hotels :")
print(city_hotels['is_canceled'].value_counts(normalize=True))
```

41.71% booikings are cancelled in City Hotels

```python
resort_hotel = resort_hotels.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotels.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title("Average Daily Rate in City and Resort Hotels",fontsize=30)
plt.plot(resort_hotel.index,resort_hotel.adr,label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel.adr,label='City Hotel')
plt.legend(fontsize=20)
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(16,10))
ax1 = sns.countplot(x='month',hue='is_canceled',data=df,palette = 'dark')
plt.title("Reservation status per Month")
plt.xlabel('Months')
plt.ylabel('number of reservations')
plt.legend(['Not Canceled','Canceled'])
```

January has most number of cancellations and August has lowest number of cancellations

```python
plt.figure(figsize=(15,8))
plt.title('ADR per month',fontsize=20)
sns.barplot('month','adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

As we can,January has the highest price and accordingly it has highest cancellation rate and vice versa for August

```python
df.head()
```

```python
canceled_data = df[df['is_canceled']==1]
top_10_country = canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title("Top 10 countries with reservations cancelled")
plt.pie(top_10_country,autopct="%.2f",labels=top_10_country.index)
plt.show()
```

Portugal has most cancellations with upto 70.07%

```python
df['market_segment'].value_counts(normalize=True)
```

47.43% of customers are coming from online travel agents

```python
canceled_data['market_segment'].value_counts(normalize=True)
```

As we can see,46.96% of cancelations are online

```python
canceled_df_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace = True)
canceled_df_adr.sort_values('reservation_status_date',inplace=True)

not_canceled_data = df[df['is_canceled']==0]
not_canceled_df_adr = not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_df_adr.reset_index(inplace = True)
not_canceled_df_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,8))
plt.title('Average Daily Rate',fontsize = 20)
plt.plot(not_canceled_df_adr['reservation_status_date'],not_canceled_df_adr['adr'],label="not canceled")
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'],label="canceled")
plt.legend(fontsize=20)
plt.show()
```
